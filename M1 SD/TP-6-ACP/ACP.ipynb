```python
from PCA_IRIS import myACP
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Visualisation des composant de l'ACP et des donnes

```python
iris = datasets.load_iris()
y = iris.target

#######################################################################
#              ACP simple 

X = iris.data

Y, _, _ = myACP(X)    

plt.subplots(1, 2, figsize=(19, 8))
plt.subplot(1, 2, 1)
plt.scatter(Y[:, 0], Y[:, 1], c=y, cmap=plt.cm.Set1,edgecolor='k')
plt.xlabel('CP 1')
plt.ylabel('CP 2')
plt.title('myACP des donnes IRIS')



plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,edgecolor='k')
plt.xlabel('CP 1')
plt.ylabel('CP 2')
plt.title('Donnes IRIS')
plt.show()
```

```python
def variance_expliquee(X):
    n = X.shape[0]
    m = X.shape[1]
    Y , val, vec = myACP(X)
    Ut = vec.T
    first_term = X@vec  #Ut @ X.T  
    second_term = np.mean(X,axis=0)@vec  #Ut@X.T
    print(first_term.shape, second_term.shape)
    J = 1/n * np.sum((X@vec - np.mean(X,axis=0)@vec)**2, axis=0)
    return J

variance_expliquee(X)
```

```python
Y , val, vec = myACP(X)

variance_expliquee = val/np.sum(val)

bar_labels = [f"C{i+1} : {round(variance_expliquee[i],2)}" for i in range(len(variance_expliquee))]
bar_colors = ['red', 'blue', 'green', 'orange']
plt.bar(['C1','C2','C3','C4'],variance_expliquee,label=bar_labels, color=bar_colors)
# plt.yscale('log')
plt.legend()
plt.title('Variance expliquees de chaque composante')

plt.show()
```
