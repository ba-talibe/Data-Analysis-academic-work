```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = 1.5*np.random.randn(100)
```

```python
x2 = 3 * x1 + 4*np.random.rand(100) 
```

```python
plt.figure()
plt.scatter(x1,x2)
```

```python
x3 = -2 * x1 + x2 + np.random.randn(100)
```

```python
import pandas as pd

df = pd.DataFrame({'X1':x1, 'X2':x2, 'X3':x3})
df
```

```python
pd.plotting.scatter_matrix(df)
```

```python
X = np.array(df)
```

```python
X.shape
```

```python
df.corr()
```

```python
X.std(axis=0)
```

```python
X1 = X/X.std(axis=0)
```

```python
X1.std(axis=0)
```

```python
df2 = pd.DataFrame({'X1':X1[:,0], 'X2':X1[:,1], 'X3':X1[:,2]})
```

```python
df2.corr()
```

```python
pd.plotting.scatter_matrix(df2)
```

```python
S = X1.T @ X1 / 100
```

```python
S
```

```python
l , V = np.linalg.eig(S)
```

```python
print(V)
```

```python
sum(V[:,1]*V[:,2])
```

```python
X2 = X1@V
```

```python
df2 = pd.DataFrame({'X1':X2[:,0], 'X2':X2[:,1], 'X3':X2[:,2]})
```

```python
df2.corr()
```
